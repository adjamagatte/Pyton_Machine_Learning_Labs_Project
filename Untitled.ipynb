```python
def createData(filename,sepp):
    """
    This function load a dataset from a csv file
    input: filename= a csv file
           sep = the data separator
    output: dataframe
    """
    df = pd.read_csv(filename.csv, sep = sepp)
    print(df.info())
    return df
```
